```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
weather=pd.read_csv(r"C:\Users\Prakash\Downloads\Test.csv")
```

```python
weather.head()
```

```python
weather.info()
```

```python
sns.pairplot(weather[['humidity', 'temperature', 'air_pollution_index']])
```

```python
sns.countplot(x='weather_type', data=weather, palette='Set2',color='orange')
plt.title('Count of Weather Types',color='orange')
plt.xlabel('Weather Type',color='orange')
plt.ylabel('Count',color='orange')
```

```python
sns.barplot(x='humidity', y='temperature', data=weather,color='blue') 
plt.title('Scatter Plot of Humidity vs Temperature',color='orange')
plt.xlabel('Humidity',color='orange')
plt.ylabel('Temperature',color='orange')
```

```python
sns.distplot(weather['humidity'],color='blue')
plt.title('Histogram of Humidity',color='orange')
plt.xlabel('Humidity',color='orange')
plt.ylabel('Frequency',color='orange')
```

```python
sns.distplot(weather['humidity'], kde=False, rug=True,color='blue')
plt.xlabel('Humidity',color='orange')
```

```python
sns.jointplot(x='humidity', y='temperature', data=weather,color='blue') 
plt.title('JoinPlot of Humidity vs Temperature',color='orange')
plt.xlabel('Humidity',color='orange')
plt.ylabel('Temperature',color='orange')
```

```python
sns.jointplot(x=weather['humidity'], y=weather['temperature'], kind="hex",color='blue')
```

```python
sns.jointplot(x=weather['humidity'], y=weather['temperature'], kind="kde",color='blue')
```

```python
sns.stripplot(x='weather_type',y='temperature',data=weather,color='blue')
plt.title('Strip Plot of Weather Type vs Temperature',color='orange')
plt.xlabel('Weather Type',color='orange')
plt.ylabel('Temperature',color='orange')
```

```python
sns.swarmplot(x=weather['humidity'],y= weather['temperature'],color='blue')
plt.title('swarmplot of Weather Type vs Temperature',color='orange')
plt.xlabel('humidity',color='orange')
plt.ylabel('temperature',color='orange')
```

```python
sns.lmplot(x="humidity", y="temperature", hue="weather_type", data=weather)
plt.title('lmplot of Weather Type vs Temperature',color='orange')
plt.xlabel('humidity',color='orange')
plt.ylabel('temperature',color='orange')
```
